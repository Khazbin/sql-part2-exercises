To get started with the exercises, go ahead and take a look at RideShareDB. The dataset was made using data obtained from the NYC Taxi and Limosine Commission or TLC. TLC tracks this data by base and even rideshare companies such as Uber and Lyft have assigned bases that each pickup must be attributed to. ## String Functions

Write a query that returns the last 3 characters of the base name in `other-FHV-services-jan-july15`. Do you see anything that is a common abbreviation at the end of business names?

INC is a common abbreviation, but it sometimes includes the . and so is NC.

```sql
select right("Base_Name", 3) from other_FHV_services_jan_aug_2015
```

The number of trips and the number of pickups are both stored as strings in `other-FHV-services-jan-july15`. This is because the data provided included spaces in some of the numbers. Write a query that returns the location of spaces in both of these columns.

```sql
select charindex(' ',"Pick_Up_Date") as "Pick Up Date Index", charindex(' ',"Number_of_Trips") as "Number of Trips Index" from other_FHV_services_jan_aug_2015
```

## Date Functions

For `other-FHV-services-jan-july15`, display the name of month of the pickup date and order in descending order. Is the result of this query what you expected?

For the most part it is. The result does start with August and sort descending, however it doesn't go past June.

```sql
-- select datepart("month","Pick_Up_Date") as "Pick Up Month" from other_FHV_services_jan_aug_2015 order by "Pick_Up_Date" desc

Select datename(month,"Pick_Up_Date") from other_FHV_services_jan_aug_2015 as [Month Name] order by "Pick_Up_Date" desc
```

Using `other-FHV-services-jan-july15`, write a query that returns the day number of the pickup date and month name and orders it in ascending order by day number.

```sql
Select datepart("day","Pick_Up_Date"), datename(month,"Pick_Up_Date") from other_FHV_services_jan_aug_2015 order by "Pick_Up_Date" asc
```
